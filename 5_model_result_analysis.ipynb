## Compute p-values and stuff for the comparison between every feature set and demographics_only, for every method and target.

```python
import scipy.stats
aggregated_regression_rows = []
for feature_set in regression_table.feature_set.unique():
    if feature_set == 'demographics_only':
        continue
    for method in regression_table.method.unique():
        for target in regression_table.target.unique():
            data_subset = regression_table.query('feature_set==@feature_set and method==@method and target==@target')
            baseline_subset = regression_table.query('feature_set=="demographics_only" and method==@method and target==@target')
            clinical_full_subset = regression_table.query('feature_set=="clinical_full" and method==@method and target==@target')
            t_test_results = scipy.stats.ttest_ind(data_subset.r2, baseline_subset.r2, alternative='greater')
            t_vs_clinical = scipy.stats.ttest_ind(data_subset.r2, clinical_full_subset.r2, alternative='greater')
            row = {'feature_set': feature_set, 'method': method, 'target': target,
                   't_pval_vs_demo': t_test_results.pvalue,
                   't_pval_vs_clinical': t_vs_clinical.pvalue}
            aggregated_regression_rows.append(row)
```

```python
regression_pval_table = pd.DataFrame(aggregated_regression_rows)
regression_pval_table['t_pval_vs_demo_corrected'] = scipy.stats.false_discovery_control(regression_pval_table.t_pval_vs_demo, method='by')
regression_pval_table['t_pval_vs_clinical_corrected'] = scipy.stats.false_discovery_control(regression_pval_table.t_pval_vs_clinical, method='by')
regression_pval_table[regression_pval_table.method == 'LassoCV']
```

```python
aggregated_classification_rows = []
feature_set_rows = []
for feature_set in classification_table.feature_set.unique():
    feature_set_row = {'feature_set': feature_set}
    for method in classification_table.method.unique():
        for target in classification_table.target.unique():
            data_subset = classification_table.query('feature_set==@feature_set and method==@method and target==@target')
            baseline_subset = classification_table.query('feature_set=="demographics_only" and method==@method and target==@target')
            clinical_full_subset = classification_table.query('feature_set=="clinical_full" and method==@method and target==@target')
            t_test_f1 = scipy.stats.ttest_ind(data_subset.f1, baseline_subset.f1, alternative='greater')
            t_test_ba = scipy.stats.ttest_ind(data_subset.balanced_acc, baseline_subset.balanced_acc, alternative='greater')
            t_vs_clinical_f1 = scipy.stats.ttest_ind(data_subset.f1, clinical_full_subset.f1, alternative='greater')
            t_vs_clinical_ba = scipy.stats.ttest_ind(data_subset.balanced_acc, clinical_full_subset.balanced_acc, alternative='greater')
            row = {'feature_set': feature_set, 'method': method, 'target': target,
                   'f1_mean': data_subset.f1.mean(),
                   'f1_demo_diff': data_subset.f1.mean() - baseline_subset.f1.mean(),
                   'f1_clinical_diff': data_subset.f1.mean() - clinical_full_subset.f1.mean(),
                   't_pval_vs_demo_f1': t_test_f1.pvalue,
                   #'u_pval_vs_demo_ba': u_test_ba.pvalue
                   't_pval_vs_clinical_f1': t_vs_clinical_f1.pvalue}
            aggregated_classification_rows.append(row)
            if method == 'LogRegLasso':
                feature_set_row[target] = data_subset.f1.median()
    feature_set_rows.append(feature_set_row)
```

## Plotting current prediction accuracy for all tasks

```python
import matplotlib
cmap = matplotlib.cm.get_cmap('Pastel1')
```

```python
cmap(3)
```

```python
# mapping of variable names to the full title that should be displayed
name_to_full_title = {'hba1c': 'HbA1C', 'gfr': 'eGFR', 'homa-ir': 'HOMA-IR',
                      'glucose': 'Glucose', 'insulin': 'Insulin', 'bmi': 'BMI',
                      'age': 'Age'}
```

```python
methods_to_include = ['LassoCV', 'ElasticNetCV']
included_tables = [demo_results, clinical_results, clinical_full_results, prot_results,
                   met_results,
                    all_clinical_results]
table_names = ['demographics_only', 'clinical_selected', 'clinical_full', 'prot', 'met', 'all']

# plots two methods side by side
def plot_save_methods(target0, methods=methods_to_include, tables=included_tables, table_names=table_names, baseline=baseline_results):
    fig, axes = plt.subplots(1, 2)
    for ax_index, (m, ax) in enumerate(zip(methods, axes)):
        xticks = []
        data_table = []
        m_name = m
        if m_name == 'RandomForestRegressor':
            m_name = 'RF_cv'
        if m_name == 'ElasticNetCV':
            m_name = 'ENCV'
        if m_name == 'ElasticNet':
            m_name = 'EN'
        for table, name in zip(tables, table_names):
            data_table.append(table[m][target0])
            xticks.append(name)
        bp = ax.boxplot(data_table, patch_artist=True)
        method_index = 0
        for i, patch in enumerate(bp['boxes']):
            method_index = (method_index + 1) % len(tables)
            patch.set(facecolor=cmap(method_index))
            ax.set_xticks(range(1, len(xticks)+1), xticks, rotation=75)
        ax.set_xlabel('Feature set')
        if ax_index == 0:
            ax.set_ylabel('r^2 score')
        ax.set_title(target0 + ' ' + m_name)
        if target0 == 'next_hba1c':
            ax.set_ylim(0.25, 0.9)
        elif target0 == 'next_gfr':
            ax.set_ylim(0.4, 1)
        elif target0 == 'd_hba1c' or target0 == 'd_gfr' or target0 == 'd_glucose':
            ax.set_ylim(0, 0.5)
        else:
            ax.set_ylim(0, 1)
        ax.grid()
    # plot baseline
        if baseline != None:
            bl_table = baseline['baseline_linear_regression'][target0]
            median = np.median(bl_table)
            quantiles = np.quantile(baseline_results['baseline'][target0], [0.25, 0.75])
            ax.axhline(np.mean(bl_table), linestyle='--')
    plt.tight_layout()
    #plt.savefig('images_2023_10_18/{0}_clinical_7.png'.format(target0))
    
# It might be better if we plotted subgraphs for the different methods (ENCV, RFR) rather than putting
# them on the same graph.
```

### Plotting function for regression R^2 (comparing two timepoints)

```python
included_tables = [demo_results, clinical_results, clinical_full_results, prot_results,
                   met_results,
                    all_clinical_results]
included_advanced_tables = [demo_advanced_models_results, clinical_advanced_models_results, 
                            clinical_full_advanced_models_results, prot_advanced_models_results,
                   met_advanced_models_results,
                    all_clinical_advanced_models_results]
table_names = ['demographics_only', 'clinical_selected', 'clinical_full', 'prot', 'met', 'all']

# plots two variables rather than two methods, side by side
def plot_save_timepoints(variables, method='LassoCV',
                        tables=included_tables, table_names=table_names, baseline=baseline_results):
    # TODO: plot multiple panes - subplots
    fig, axes = plt.subplots(1, 2)
    for ax_index, (target0, ax) in enumerate(zip(variables, axes)):
        xticks = []
        data_table = []
        m = method
        m_name = method
        for table, name in zip(tables, table_names):
            data_table.append(table[m][target0])
            xticks.append(name)
        bp = ax.boxplot(data_table, patch_artist=True)
        method_index = 0
        for i, patch in enumerate(bp['boxes']):
            method_index = (method_index + 1) % len(tables)
            patch.set(facecolor=cmap(method_index))
            ax.set_xticks(range(1, len(xticks)+1), xticks, rotation=75)
        ax.set_xlabel('Feature set')
        if ax_index == 0:
            ax.set_ylabel('r^2 score')
        if '1y' in target0:
            ax.set_title('1 year')
        else:
            ax.set_title('6 months')
        ax.set_ylim(0, 1)
        ax.grid()
    # plot baseline
        if baseline != None:
            bl_table = baseline['baseline_linear_regression'][target0]
            median = np.median(bl_table)
            quantiles = np.quantile(baseline_results['baseline'][target0], [0.25, 0.75])
            ax.axhline(np.mean(bl_table), linestyle='--')
    fig.suptitle('{0} regression'.format(name_to_full_title[variables[0].split('_')[1]]))
    plt.tight_layout()
    if method != 'LassoCV':
        plt.savefig('images_2023_10_18/{0}_{1}_regression.png'.format(variables[0], method), dpi=300)
    else:
        plt.savefig('images_2023_10_18/{0}_regression.png'.format(variables[0]), dpi=300)
```

```python
f1_class_tables = [demo_results_class_f1, clinical_results_class_f1, 
                   clinical_full_results_class_f1, prot_results_class_f1, met_results_class_f1,
                    all_clinical_results_class_f1]
f1_advanced_class_tables = [demo_advanced_models_results_class_f1, clinical_advanced_models_results_class_f1, 
                   clinical_full_advanced_models_results_class_f1, prot_advanced_models_results_class_f1,
                            met_advanced_models_results_class_f1,
                    all_clinical_advanced_models_results_class_f1]
def plot_save_class_f1(variables,
                       method='LogRegLassoCV',
                       tables=f1_class_tables, table_names=class_table_names, baseline=None):
    # TODO: plot multiple panes - subplots
    fig, axes = plt.subplots(1, 2)
    for ax_index, (target0, ax) in enumerate(zip(variables, axes)):
        xticks = []
        data_table = []
        m = method
        m_name = m
        if m_name == 'RandomForestRegressor':
            m_name = 'RF'
        if m_name == 'ElasticNetCV':
            m_name = 'LogReg-ENCV'
        if m_name == 'LassoCV':
            m_name = 'LogReg-LassoCV'
        for table, name in zip(tables, table_names):
            data_table.append(table[m][target0])
            xticks.append(name)
        bp = ax.boxplot(data_table, patch_artist=True)
        method_index = 0
        for i, patch in enumerate(bp['boxes']):
            method_index = (method_index + 1) % len(tables)
            patch.set(facecolor=cmap(method_index))
            ax.set_xticks(range(1, len(xticks)+1), xticks, rotation=75)
        ax.set_xlabel('Feature set')
        if ax_index == 0:
            ax.set_ylabel('f1 score')
        if '1y' in target0:
            ax.set_title('1 year')
        else:
            ax.set_title('6 months')        
        ax.set_ylim(0, 1)
        ax.grid()
    # TODO: p-values?
    title_base = name_to_full_title[variables[0].split('_')[2]]
    fig.suptitle('Change in ' + title_base)
    plt.tight_layout()
    if method != 'LogRegLassoCV':
        plt.savefig('images_2023_10_18/{0}_{1}_f1.png'.format(variables[0], method), dpi=300)
    else:
        plt.savefig('images_2023_10_18/{0}_f1.png'.format(variables[0]), dpi=300)
```
